# Incorporation of PSO into first-principles methods for structural optimization of twisted bilayer graphene

## Objective function

We use the Lennard-Jones potential to compute the energy $E$ of $N$ atoms:
  pairwise energies given by
  \begin{equation*}
    E = 4 \epsilon \sum\limits_{i=1}^N \sum\limits_{j=i+1}^N \left(
    \left(\frac{\sigma}{r_{ij}}\right)^{12} - \left(\frac{\sigma}{r_{ij}}\right)^6 \right),
  \end{equation*}
  where $r_{ij}$ is the distance between atom $i$ and atom $j$. For graphene, $\epsilon = 2.96$ meV and $\sigma=3.4$ Angstroms.

```python
# Define lattice constants for the graphene unit cell and C-C bond length
a1 = 2.46 * np.array([1, 0])
a2 = 2.46 * np.array([1 / 2, np.sqrt(3) / 2])
cc = (a1 + a2) / 3

# Constants for graphene in the L-J potential
sigma = 3.4
eps = 2.96
```

```python
# Functions for calculating twist angles and number of atoms for a value of N.
# We define N as follows: A moire supercell is described by lattice vectors
# equal to N * a1 and (N - 1) * a2, where a1 and a2 are the primitive 
# lattice vectors defined above

def angles(N):
    return np.round(
        np.arccos(
            ((N - 1) ** 2 + 4 * N * (N - 1) + N ** 2) / (2 * ((N - 1) ** 2 + N * (N - 1) + N ** 2))
        ) * 180 / np.pi, 
        10
    )

def num_atoms(N):
    return int(np.round(1 / np.square(np.sin(np.pi / 180 * angles(N) / 2)), 0))
```

## PSO implmementation for AA and AB-stacked graphene

```python
### PSO for AA and AB-stacked graphene ###

# Objective function
# z = array of length 2 containicng the z-coordinate of the 2 layers
# AB = flag for doing PSO for AB-stacked graphene (by default, we do AA stacked)
def fun_no_twist(z, AB=False):
    all_pos = []
    E = 0

    # Generate a 9x9 array of unit cells
    for idx, xi in enumerate(z):
        for i in [-4, -3, -2, -1, 0, 1, 2, 3, 4]:
            for j in [-4, -3, -2, -1, 0, 1, 2, 3, 4]:
                for c in [0, 1]:
                    pos = [np.array([c * cc[0] + i * a1[0] + j * a2[0], c * cc[1] + i * a1[1] + j * a2[1], xi])]

                    # Shift one layer if doing AB-stacking
                    if AB and idx == 1:
                        pos = [np.array([pos[0][0] + cc[0], pos[0][1] + cc[1], pos[0][2]])]
                    all_pos += pos

    # Calculate energy under the L-J potential
    for i, pos_i in enumerate(all_pos[:len(all_pos) // 2]):
        for j, pos_j in enumerate(all_pos[len(all_pos) // 2:]):
            r2 = np.linalg.norm(pos_i - pos_j) ** 2
            r2inv = sigma ** 2 / (r2 + 1e-16)  # 3.4 for graphene
            r6inv = r2inv ** 3
            r12inv = r6inv ** 2
            E += r12inv - r6inv
    return eps * E

# PSO algorithm
# lo, hi = lower and upper limits for the atomic positions
# nstep = number of steps
# AB = flag for doing PSO for AB-stacked graphene (by default, we do AA stacked)
# S = Swarm size
def pso_no_twist(lo, hi, nstep, AB=False, S=100):
    D = len(lo)  # Number of degrees of freedom
    
    # x is a list of shape (S, size), where S is the number of particles and 
    # each particle is a configuration of D (size) atomic positions. Same for v,
    # but v is a velocity
    x = [[(random.uniform(lo, hi)) for lo, hi in zip(lo, hi)] for i in range(S)]  # Randomly initialize positions
    v = [[(random.uniform(lo, hi) - x) / 2 for x, lo, hi in zip(x[i], lo, hi)]  # Randomly initialize velocities
         for i in range(S)]
    p = copy.deepcopy(x)

    # score is a list of length S containing the energy for each particle (configuration)
    score = [fun_no_twist(x, AB) for x in x]

    # globally optimal score (energy) and positions found so far
    best_idx = np.argmin(score)
    sg, pg = score[best_idx], p[best_idx]
    # print(sg)
    # print(pg)
    alpha = 0.9
    c = 1
    for istep in range(nstep):
        for i in range(S):
            for d in range(D):
                # PSO update step
                v[i][d] = (alpha - (alpha - 0.4) * istep / nstep) * v[i][d] + random.uniform(0, 
                        c) * (pg[d] - x[i][d]) + random.uniform(
                        0, c) * (p[i][d] - x[i][d])
                x[i][d] += v[i][d]
                # Don't let the coordinates go out of bounds
                if x[i][d] < lo[d]:
                    x[i][d] = lo[d]
                    v[i][d] = 0
                if x[i][d] > hi[d]:
                    x[i][d] = hi[d]
                    v[i][d] = 0
        for i in range(S):
            s = fun_no_twist(x[i], AB)
            # If new score is better than current score for this particle, replace the score (and replace the particle's positions too)
            if s < score[i]:
                score[i] = s
                p[i] = x[i][:]
                # If new score is better than the globally optimal score, replace (and replace the globally optimal position)
                if s < sg:
                    sg = s
                    pg = x[i][:]
        # print(sg)
    return sg, pg
```

## PSO-1 implementation

```python
### PSO-1 (optimizing average bilayer spacing) ###

# Objective function
# z = list of the z-coordinates of each layer
# struct = original structure
def fun_2params(z, struct): 
    x = []

    # Construct 3D positions using optimized z coordinate of each layer and original xy coord of each atom
    for i in range(len(struct) // 2):
        x.append(np.array([struct[i, 0], struct[i, 1], z[0]]))
    for i in range(len(struct) // 2, len(struct)):
        x.append(np.array([struct[i, 0], struct[i, 1], z[1]]))
    
    E = 0

    # Add L-J energy between the layers
    for i, pos_i in enumerate(x[:len(x) // 2]):
        for j, pos_j in enumerate(x[len(x) // 2:]):
            r2 = np.linalg.norm(pos_i - pos_j) ** 2
            r2inv = sigma ** 2 / (r2 + 1e-16)
            r6inv = r2inv ** 3
            r12inv = r6inv ** 2
            E += r12inv - r6inv
    return eps * E
    
# PSO algorithm
# lo, hi = lower and upper limits for the atomic positions
# nstep = number of steps
# initial_structure = original structure
# S = Swarm size
def pso_twist_2params(lo, hi, nstep, initial_structure, S=100):
    D = len(lo)  # Number of degrees of freedom
    
    # x is a list of the z coordinate of each layer (confusing naming but I'm too lazy to change it)
    x = [[initial_structure[0, 2], initial_structure[-1, 2]] for i in range(S)]
    
    v = [[-(random.uniform(0, np.amax(x)) - x_i) / 200 for x_i, lo, hi in zip(x[i], lo, hi)]  # Randomly initialize velocities
     for i in range(S)]
    p = copy.deepcopy(x)

    # score is a list of length S containing the energy for each particle (configuration)
    score = [fun_2params(x, initial_structure) for x in x]
    
    # globally optimal score (energy) and positions found so far
    best_idx = np.argmin(score)
    sg, pg = score[best_idx], p[best_idx]
    # print(sg)
    # print(pg)
    alpha = 0.9
    c = 1
    for istep in range(nstep):
        for i in range(S):
            for d in range(D):
                # PSO update step
                v[i][d] = (alpha - (alpha - 0.4) * istep / nstep) * v[i][d] + random.uniform(0, 
                        c) * (pg[d] - x[i][d]) + random.uniform(
                        0, c) * (p[i][d] - x[i][d])
                x[i][d] += v[i][d]
                # Don't let the coordinates go out of bounds
                if x[i][d] < lo[d]:
                    x[i][d] = lo[d]
                    v[i][d] = 0
                if x[i][d] > hi[d]:
                    x[i][d] = hi[d]
                    v[i][d] = 0
        for i in range(S):
            s = fun_2params(x[i], initial_structure)
            # If new score is better than current score for this particle, replace the score (and replace the particle's positions too)
            if s < score[i]:
                score[i] = s
                p[i] = x[i][:]
                # If new score is better than the globally optimal score, replace (and replace the globally optimal position)
                if s < sg:
                    sg = s
                    pg = x[i][:]
        # print(sg)
    return sg, pg
```

## PSO-2 implementation

```python
### PSO-2 (optimizing each atomic z-coordinate independently) ###

# Objective function
# z = list of the z-coordinates of all atoms
# struct = original structure
# neighbors = list of atoms bonded to each other
def fun_twist(z, struct, neighbors): 

    # Construct 3D positions using optimized z coordinates and original xy coords
    x = [np.array([struct[i, 0], struct[i, 1], z_i]) for i, z_i in enumerate(z)]
    E = 0

    # Add C-C bond stretching energy
    for i, pos_i in enumerate(x):
        for neighbor in neighbors[i]:
            E += 0.5 * 300000 * (1.420281662 - np.linalg.norm(x[neighbor] - pos_i)) ** 2

    # Add L-J energy between the layers
    for i, pos_i in enumerate(x[:len(x) // 2]):
        for j, pos_j in enumerate(x[len(x) // 2:]):
            r2 = np.linalg.norm(pos_i - pos_j) ** 2
            r2inv = sigma ** 2 / (r2 + 1e-16)
            r6inv = r2inv ** 3
            r12inv = r6inv ** 2
            E += r12inv - r6inv
    return eps * E

# PSO algorithm
# lo, hi = lower and upper limits for the atomic positions
# nstep = number of steps
# initial_structure = original structure
# neighbors = list of atoms bonded to each other
# S = Swarm size
def pso_twist(lo, hi, nstep, initial_structure, neighbors, S=100):
    D = len(lo)  # Number of degrees of freedom
    
    # x is a list of the z-coordinates of the atoms (confusing naming but I'm too lazy to change it)
    x = []
    for i in range(S):
        x.append(initial_structure[:, 2])
    v = [[-(random.uniform(0, np.amax(x)) - x_i) / 200 for x_i, lo, hi in zip(x[i], lo, hi)]  # Randomly initialize velocities
         for i in range(S)]
    p = copy.deepcopy(x)

    # score is a list of length S containing the energy for each particle (configuration)
    score = [fun_twist(x, initial_structure, neighbors) for x in x]
    
    # globally optimal score (energy) and positions found so far
    best_idx = np.argmin(score)
    sg, pg = score[best_idx], p[best_idx]
    # print(sg)
    # print(pg)
    alpha = 0.9
    c = 1
    for istep in range(nstep):
        for i in range(S):
            for d in range(D):
                # PSO update step
                v[i][d] = (alpha - (alpha - 0.4) * istep / nstep) * v[i][d] + random.uniform(0, 
                        c / 10) * (pg[d] - x[i][d]) + random.uniform(
                        0, c) * (p[i][d] - x[i][d])
                x[i][d] += v[i][d]
                # Don't let the coordinates go out of bounds
                if x[i][d] < lo[d]:
                    x[i][d] = lo[d]
                    v[i][d] = 0
                if x[i][d] > hi[d]:
                    x[i][d] = hi[d]
                    v[i][d] = 0
        for i in range(S):
            s = fun_twist(x[i], initial_structure, neighbors)
            # If new score is better than current score for this particle, replace the score (and replace the particle's positions too)
            if s < score[i]:
                score[i] = s
                p[i] = x[i][:]
                # If new score is better than the globally optimal score, replace (and replace the globally optimal position)
                if s < sg:
                    sg = s
                    pg = x[i][:]
        # print(sg)
    return sg, pg
```

```python
# Function for extracting the runtimes and number of geometry optimization steps in the DFT relaxation procedure
def count_word_and_last_word_in_clock_files(base_folder, word):
    """
    Count the occurrences of a word and get the last word of the last line in files named 'CLOCK'.
    The last word is the total run time of the geometry optimization procedure in DFT.

    Parameters:
        base_folder (str): The base directory to start searching for files.
        word (str): The word to count in the files.
    """
    counts = []
    times = []
    
    # Walk through all directories in the base_folder
    for root, dirs, files in os.walk(base_folder):
        if 'CLOCK' in files:
            # Construct the file path
            file_path = os.path.normpath(os.path.join(root, 'CLOCK'))
            
            try:
                # Read the file and count occurrences
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read()
                    count = content.count(word)
                
                # Find the last word of the last line
                with open(file_path, 'r', encoding='utf-8') as file:
                    lines = file.readlines()
                    if lines:  # Check if the file is not empty
                        last_line = lines[-1].strip()
                        last_word = last_line.split()[-1] if last_line else "(no last word)"
                    else:
                        last_word = "(empty file)"
                
                # Print results
                # print(f"File: {file_path}, Word: '{word}', Count: {count}, Last Word: '{last_word}'")
                counts.append(int(count))
                times.append(float(last_word))
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")
    return counts, times

# Extract 
counts, times = [], []
for N in [4, 5, 6, 7, 8, 9]:
    base_folder = fr"C:\Users\alber\AM207-final-project\N_{N}"  # Use absolute path if on Windows. Replace with your folder path on a normal computer
    word_to_search = "geometry"  # The number of times this work appears is the number of geometry optimization steps
    count, time = count_word_and_last_word_in_clock_files(base_folder, word_to_search)
    if N == 4:
        count = count[::2]
        time = time[::2]
    counts.append(count)
    times.append(time)
counts = np.array(counts)
times = np.array(times)
```
